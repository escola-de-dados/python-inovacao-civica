### Leia o csv e converta para dicionário

```python
def read_file(file):
    with open(f'{file}.csv', 'r', encoding='latin_1') as candidatos:
        csv_linhas = reader(candidatos)
        lista = [lista  for lista in csv_linhas]
    
        keys = lista.pop(0)
        fill = keys[0].replace('"', "")

        keys = fill.split(';')
        
        lista_candidatos = []
        for i in range(len(lista)):
            cad = lista[i][0].replace('"',"")
            filter_cad = cad.split(';')
            cand = dict(zip(keys, filter_cad))
            lista_candidatos.append(cand)

    return lista_candidatos
    
candidatos = read_file('consulta_cand_2020_GO')


candidatos
```
