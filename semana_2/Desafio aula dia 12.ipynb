```python
import pandas as pd
import glob
```

```python
lista_arquivos = glob.glob('conjunto_dados/*.csv')
print(lista_arquivos)
```

```python
lista_df = []
```

```python
for csv in lista_arquivos:
    df = pd.read_csv(csv, sep=';', encoding='latin1')
#     print(df.info())
    lista_df.append(df)
```

```python
candidaturas_PE_RN = pd.concat(lista_df, axis=0, ignore_index=True)
```
